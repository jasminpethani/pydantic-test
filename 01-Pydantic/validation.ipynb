```python
from pydantic import BaseModel, Field

class Item(BaseModel):
    name:str=Field(min_length=8, max_length=20)
    department:str=Field(default="IT", description="this department of item prop")
    

try: 
    item=Item(name="jasminpethani")
except ValueError as e: 
    print(e) 
else: 
    print(item)

```
