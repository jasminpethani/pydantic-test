```python
from typing import Optional 

class Employee(BaseModel):
    id:int
    name:str
    department:str
    salary: Optional[float] = None # Optional here is default value if nothing passed 
    is_active: Optional[bool] = True


emp=Employee(id=1, name="jasmin", department="IT")
print(emp)
```

```python
## auto type casting

emp1=Employee(id=1, name="jasmin", department="IT", salary=12000, is_active=False)
print(emp1)
```
